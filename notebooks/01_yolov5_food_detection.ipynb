```python
img_list = glob('./dataset/images/*.jpg')
len(img_list)
```

```python
from sklearn.model_selection import train_test_split
# train test split
train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=129)

print(len(train_img_list), len(val_img_list))
```

```python
# yolov5 경로
# if not os.path.isdir('../archive/raw-data/yolo/dataset'):
#     os.makedirs('../archive/raw-data/yolo/dataset')
```

```python
# train test split하고 난 파일들의 경로를 저장하고 있는 txt 파일
with open('../archive/raw-data/대외/2023_핀테크_공모전/IMG/train.txt', 'w') as f :
  f.write('\n'.join(train_img_list) + '\n')

with open('../archive/raw-data/대외/2023_핀테크_공모전/IMG/val.txt', 'w') as f:
  f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml

with open('../archive/raw-data/대외/2023_핀테크_공모전/IMG/dataset/data.yaml', 'r') as f:
  data = yaml.load(f)

print(data)

data['train'] = '../archive/raw-data/대외/2023_핀테크_공모전/IMG/train.txt' # yolov5 경로
data['val'] = '../archive/raw-data/대외/2023_핀테크_공모전/IMG/val.txt' # yolov5 경로

with open('../archive/raw-data/대외/2023_핀테크_공모전/IMG/dataset/data.yaml', 'w') as f:
  yaml.dump(data, f)
print(data)
```
