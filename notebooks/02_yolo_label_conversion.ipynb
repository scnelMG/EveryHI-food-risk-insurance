```python
import torch
from PIL import Image
import torchvision.transforms as transforms
from torch.backends import cudnn

# GPU 가속을 활성화합니다.
cudnn.benchmark = True

# YOLOv5 .pt 파일 경로를 지정합니다.
weights_path = "/content/final_best.pt"

# 모델을 로드합니다.
model = torch.hub.load('ultralytics/yolov5', 'custom', path=weights_path)

# 이미지를 불러올 경로와 파일명을 지정합니다.
image_path = "/content/KakaoTalk_20230716_152751815.png"

# 이미지를 PIL 형식으로 열고 정규화합니다.
image = Image.open(image_path)
image = transforms.ToTensor()(image)

# 이미지를 모델에 입력하기 위해 차원을 추가합니다.
image = image.unsqueeze(0)

# GPU를 사용할 경우, 이미지를 GPU로 이동합니다.
if torch.cuda.is_available():
    model = model.cuda()
    image = image.cuda()

# 추론을 수행합니다.
results = model(image)

# 결과에서 라벨을 추출합니다.
labels = results.pandas().xyxy[0]['name']

# 라벨을 출력합니다.
print(labels)
```

```python
import torch
from PIL import Image
import torchvision.transforms as transforms
import cv2

# YOLOv5 .pt 파일 경로를 지정합니다.
weights_path = "/content/final_best.pt"

# 모델을 로드합니다.
model = torch.hub.load('ultralytics/yolov5', 'custom', path=weights_path)

image_path = "/content/KakaoTalk_20230716_152751815.png"

# 이미지를 열고 채널 수를 확인합니다.
image = Image.open(image_path)
channels = image.split()

# 이미지가 RGBA 형식인 경우 알파 채널을 제거합니다.
if len(channels) == 4:
    image = image.convert("RGB")


# 이미지의 채널 수를 확인합니다.
print(image.mode)  # 출력: RGB

# 추론을 수행합니다.
results = model(image)

# 결과에서 라벨을 추출합니다.
labels = results.pandas().xyxy[0]['name']

# 라벨을 출력합니다.
print(labels)

# 결과에서 라벨과 박스를 추출합니다.
labels = results.pandas().xyxy[0]['name']
boxes = results.pandas().xyxy[0][['xmin', 'ymin', 'xmax', 'ymax']].values

# 이미지를 OpenCV 형식으로 변환합니다.
image_cv = cv2.cvtColor(cv2.imread(image_path), cv2.COLOR_BGR2RGB)

# 박스와 라벨을 이미지에 추가합니다.
for box, label in zip(boxes, labels):
    x1, y1, x2, y2 = box.astype(int)
    image_cv = cv2.rectangle(image_cv, (x1, y1), (x2, y2), (100, 0, 100), 4)
    image_cv = cv2.putText(image_cv, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

output_path = "/content/output.jpg"  # 저장할 이미지의 경로와 파일명을 지정합니다.
cv2.imwrite(output_path, cv2.cvtColor(image_cv, cv2.COLOR_RGB2BGR))
print("Image saved successfully.")
```
